#### 2. Write a Python program to input marks in 3 subjects; compute average and then calculate grade as per following guidelines

```python
subject1 = int (input 
                ("please input ur first mark for the first subject: "))
subject2 = int (input 
                ("please input ur second mark for the second subject: "))
subject3 = int (input 
                ("please input ur third mark for the third subject: "))
average = (subject1 + subject2 + subject3) / 3
print ("Your average mark is : ",
       average)

guidelines = ((subject1 + subject2 + subject3)
       /300)*100

if guidelines >= 90 :
    print("Delimitation: Greater than or equal to 90")
    print ("Grade: A")
    
elif guidelines <= 89 and guidelines >= 80 :
    print("Delimitation: 89 to 80")
    print ("Grade: B")
    
elif guidelines <= 79 and guidelines >= 60 :
    print("Delimitation: 79 to 60")
    print ("Grade: C")
    
elif guidelines <= 59 and guidelines >= 50 :
    print("Delimitation: 59 to 50")
    print ("Grade: D")
    
elif guidelines <= 49 and guidelines >= 40 :
    print("Delimitation: 49 to 40")
    print ("Grade: E")
    
elif guidelines <= 39 :
    print("Delimitation: Less than or equal to 39")
    print ("Grade: F")
```

## Problem 4 Web Scraping Problem
1. Scrap the table titled: ‚Äú10 most populous countries‚Äù shown below and create a DataFrame named DF_Pop 

```python
from bs4 import BeautifulSoup
import requests 
```

```python
url = 'https://en.wikipedia.org/wiki/World_population'
```

```python
d  = requests.get(url).text
```

```python
s = BeautifulSoup(d,"html.parser")
```

```python
t = s.find_all('table')
t
```

```python
len(t) #Checking out table length
```

```python
tables = 3
```

```python
DF_Pop = pd.DataFrame(columns=["Country/Dependency", "Population", "Date", "Source(official or from the United Nations)"])

for row in t[tables].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        country = col[0].text.strip()
        pop = col[1].text.strip()
        perctentage = col[2]
        date = col[3].text
        source = col[4].text
        DF_Pop = DF_Pop.append({ "Country/Dependency":country, "Population":pop, "Date":date, "Source(official or from the United Nations)":source}, ignore_index=True)
        DF_Pop.index.name = 'Rank'
        DF_Pop.index += 1

DF_Pop
```
